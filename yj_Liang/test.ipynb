```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
y, sr = librosa.load(librosa.util.example_audio_file(), sr=44100)
```

```python
S = librosa.stft(y, n_fft=1024, hop_length=512)

print(sr)
mel_basis = librosa.filters.mel(sr, n_fft=1024, n_mels=128)
mel_S = np.dot(mel_basis, np.abs(S))
mel_S = np.log10(1+10*mel_S)
#mel_S = mel_S.T

print(mel_S.shape)
plt.figure(figsize=(10,4))
librosa.display.specshow(librosa.power_to_db(mel_S, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
```
